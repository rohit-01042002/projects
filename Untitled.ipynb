An algorithm to predict number of delayed flights

```python
numeric_features
```

```python
data=df[numeric_features].dropna()
```

```python
data.head()
```

```python
x=data.drop(['arr_delay','arr_cancelled'],axis=1)
y=data['arr_delay']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=42)
```

```python
from sklearn.preprocessing import RobustScaler
scaler=RobustScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(x_train_scaled,y_train)
```

```python
y_pred=regressor.predict(x_test_scaled)
```

```python
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test,y_pred)
print(mae)
```

```python
np.mean(y_test)
```

```python
np.mean(y_pred)
```

```python
np.max(abs(y_test-y_pred))
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(9,6))
plt.scatter(x_test['carrier_delay'],y_test,color='red')
plt.scatter(x_test['carrier_delay'],y_pred,color='green')
```

```python
#Random forest
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=40)
model.fit(x_train_scaled,y_train)
```

```python
y_pred=model.predict(x_test_scaled)
```

```python
mae=mean_absolute_error(y_test,y_pred)
print(mae)
```

```python
plt.figure(figsize=(9,6))
plt.scatter(x_test['carrier_delay'],y_test,color='red')
plt.scatter(x_test['carrier_delay'],y_pred,color='green')
```
