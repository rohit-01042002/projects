```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('US_Regional_Sales_Data.csv')
```

```python
df.shape
```

```python
df.info()
```

```python
from datetime import datetime
date_object=['ProcuredDate','OrderDate','ShipDate']
for d in date_object:
    df[d]=df[d].apply(lambda x: datetime.strptime(x,"%d/%m/%y"))
```

```python
cost=['Unit Cost','Unit Price']
for c in cost:
    df[c]=df[c].apply(lambda x: ''.join(x.split(',')))
    df[c]=df[c].apply(lambda x: float(x))
```
